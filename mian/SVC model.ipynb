```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop("class", axis=1),df['class'],test_size=0.20)
```

```python
scaled_features = scaler.transform(X_train)
scaled_valid = scaler.transform(X_test)
scaled_test = scaler.transform(test)
```

```python
df_train = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_val = pd.DataFrame(scaled_valid,columns=df.columns[:-1])
df_test = pd.DataFrame(scaled_test,columns=df.columns[:-1])
df_train.head()
```
